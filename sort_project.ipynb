```python
import time
import numpy as np
import matplotlib.pyplot as plt
import math
#import tracemalloc

def partition(arr, low, high):
    i = (low-1)         # index of smaller element
    pivot = arr[high]     # pivot   Use the last element for comparsion
    for j in range(low, high):
        if arr[j] <= pivot:
            i = i+1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)
 
def quickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
        
def insertionSort(arr, n): 
    for i in range(1, n): 
        key = arr[i] 
        j = i-1
        while j >=0 and key < arr[j] : 
                arr[j+1] = arr[j] 
                j -= 1
        arr[j+1] = key 

def selectionSort(arr, n):
    for i in range(n):
        min_idx = i 
        for j in range(i+1, len(arr)): 
            if arr[min_idx] > arr[j]: 
                min_idx = j   
        arr[i], arr[min_idx] = arr[min_idx], arr[i] 

def bubbleSort(arr, n):
    for i in range(n-1): 
        for j in range(0, n-i-1): 
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 

def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        mergeSort(L)
        mergeSort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    
runTimes=5
pointNum=5
arr = []
oriArr=[]
sortedness=[[981, 1964, 2925, 3913, 4892],
[985, 1954, 2924, 3896, 4873],
[1012, 1998, 2976, 3980, 4981],
[980, 1991, 3004, 4011, 5021]]

datasetNum = 1

datasetFile="dataset"+str(datasetNum)+".txt"
n = 10000
f = open(datasetFile, "r")
for i in range(0,n):
    arr.append(float(f.readline()))
    oriArr.append(float(f.readline()))
mergeTime=[]
insertTime=[]
bubbleTime=[]
quickTime=[]
selectTime=[]
mergeMem=[]
quickMem=[]
selectMem=[]
insertMem=[]
bubbleMem=[]
tsumMerge = 0.0
tsumQuick = 0.0
tsumInsert = 0.0
tsumSelect = 0.0
tsumBubble = 0.0
mMerge = 0.0
mQuick = 0.0
mInsert = 0.0
mSelect = 0.0
mBubble = 0.0
x1=[]
values={"merge":[], "quick":[], "bubble":[], "insert":[], "select":[]  }
sd={"merge":[], "quick":[], "bubble":[], "insert":[], "select":[]  }
for i in range(pointNum):
    b = int(n / pointNum * (i+1))
    x1.append(b)
    tsumMerge = 0.0
    tsumQuick = 0.0
    tsumInsert = 0.0
    tsumSelect = 0.0
    tsumBubble = 0.0
    mMerge = 0.0
    mQuick = 0.0
    mInsert = 0.0
    mSelect = 0.0
    mBubble = 0.0
    values["merge"]=[]
    values["quick"]=[]
    values["insert"]=[]
    values["select"]=[]
    values["bubble"]=[]
    for j in range(runTimes):
    
        # Merge Sort
        time1 = time.perf_counter()
        #tracemalloc.start()
        mergeSort(arr[:b])
        #current, peak = tracemalloc.get_traced_memory()
        #tracemalloc.stop()
        #mMerge += peak/10**6
        time2 = time.perf_counter()
        tsumMerge += time2 - time1
        values["merge"].append(time2 - time1)
        for k in range(b):
            arr[k]=oriArr[k]
    
        # Quicksort
        time1 = time.perf_counter()
        #tracemalloc.start()
        quickSort(arr, 0, b-1)
        #current, peak = tracemalloc.get_traced_memory()
        #tracemalloc.stop()
        #mQuick += peak/10**6
        time2 = time.perf_counter()
        tsumQuick += time2 - time1
        values["quick"].append(time2 - time1)
        for k in range(b):
            arr[k]=oriArr[k]
            
        # Bubble Sort
        time1 = time.perf_counter()
        #tracemalloc.start()
        bubbleSort(arr, b)
        #current, peak = tracemalloc.get_traced_memory()
        #tracemalloc.stop()
        #mBubble += peak/10**6
        time2 = time.perf_counter()
        tsumBubble += time2 - time1
        values["bubble"].append(time2 - time1)
        for k in range(b):
            arr[k]=oriArr[k]
            
        # Insertion Sort
        time1 = time.perf_counter()
        #tracemalloc.start()
        insertionSort(arr, b)
        #current, peak = tracemalloc.get_traced_memory()
        #tracemalloc.stop()
        #mInsert += peak/10**6
        time2 = time.perf_counter()
        tsumInsert += time2 - time1
        values["insert"].append(time2 - time1)
        for k in range(b):
            arr[k]=oriArr[k]
            
        # Selection Sort
        time1 = time.perf_counter()
        #tracemalloc.start()
        selectionSort(arr, b)
        #current, peak = tracemalloc.get_traced_memory()
        #tracemalloc.stop()
        #mSelect += peak/10**6
        time2 = time.perf_counter()
        tsumSelect += time2 - time1
        values["select"].append(time2 - time1)
        for k in range(b):
            arr[k]=oriArr[k]
    mergeTime.append(math.log10(tsumMerge / runTimes))
    quickTime.append(math.log10(tsumQuick / runTimes))
    insertTime.append(math.log10(tsumInsert / runTimes))
    selectTime.append(math.log10(tsumSelect / runTimes))
    bubbleTime.append(math.log10(tsumBubble / runTimes))
    # 算标准差
    
    sd["merge"].append(np.std(np.array(values["merge"])))
    sd["quick"].append(np.std(np.array(values["quick"])))
    sd["insert"].append(np.std(np.array(values["insert"])))
    sd["bubble"].append(np.std(np.array(values["bubble"])))
    sd["select"].append(np.std(np.array(values["select"])))
    
    '''mergeMem.append(mMerge/runTimes)
    quickMem.append(mQuick/runTimes)
    insertMem.append(mInsert/runTimes)
    selectMem.append(mSelect/runTimes)
    bubbleMem.append(mBubble/runTimes)'''
    
#print(mergeMem) # MB
'''print(mergeTime)
print(quickTime)'''

print(sd)

plt.xlabel('Sortedness')
plt.ylabel('Log10(running time/s)')
plt.plot(sortedness[datasetNum-1], mergeTime, sortedness[datasetNum-1], quickTime, sortedness[datasetNum-1], insertTime, sortedness[datasetNum-1], selectTime, sortedness[datasetNum-1], bubbleTime)
plt.show()
```
